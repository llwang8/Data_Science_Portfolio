#  Explore dataset on births in the U.S. from 1994 to 2003

The dataset on births in the U.S. was compiled by FiveThirtyEight.

The dataset contains the following columns:

year: Year (1994 to 2003).
month: Month (1 to 12).
date_of_month: Day number of the month (1 to 31).
day_of_week: Day of week (1 to 7).
births: Number of births that day.

This project aims to use python programming skills to analyze data and display total births by year, month, week, day of the month and day of the week.


### Read in data file and Change data type to interger

```python
import matplotlib.pyplot as plt
from numpy import arange
%matplotlib inline

data = open('US_births_1994-2003_CDC_NCHS.csv', 'r').read().split('\n')
print(data[0:10])
```

```python
# Create a function to read in string data from csv file and 
# and convert each field to integer and return a list of lists 
# of intergers

def read_csv(file):
    string_list = open(file, 'r').read().split('\n')[1:]
    final_list = []
    for yr in string_list:
        int_fields = []
        string_fields = yr.split(',')
        int_fields = [int(x) for x in string_fields]
        final_list.append(int_fields)
    return final_list

# CDC data of births from 1994 to 2003
cdc_list = read_csv('US_births_1994-2003_CDC_NCHS.csv')

print(cdc_list[0:10])    
```

### Total number of births by month over the year 1994 to 2003

```python
# calculate the total number of births that occured in
# each month, across all of the years in the dataset.

def month_births(blist):
    births_per_month = {}
    for yr in blist:
        month = yr[1]
        births = yr[4]
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
    return births_per_month

# cdc births dataset: births by month over 1994 to 2003
cdc_month_births = month_births(cdc_list)
cdc_month_births
    
```

### Total number of births by days of the week over the years 1994 to 2003

```python
# Create a function that calculates the total number of
# births for each unique day of the week

def dow_births(blist):
    result = {}
    for row in blist:
        if row[3] in result:
            result[row[3]] += row[4]
        else:
            result[row[3]] = row[4]
    return result

# cdc births dataset: births by day of the week
cdc_day_births = dow_births(cdc_list)
cdc_day_births
```

### A function to tally births by any column, be it month, day of the month, day of the week

```python
# create a single function that works for any column 
# to tally births by.

def calc_counts(data, column):
    result = {}
    for row in data:
        if row[column] in result:
            result[row[column]] += row[4]
        else:
            result[row[column]] = row[4]
    return result

# Apply function on cdc births data
cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3)

cdc_month_births
```

```python
# Create a function to transform a dict to list and 
# display the data
def transform_display(d_dict, c, label):
    length = len(d_dict)
    bar_positions = arange(length) + .75
    #print(bar_positions)
    bar_heights = list(d_dict.values())
    #print(bar_heights)
    tick_position = range(1, length+1)
    r = 30
    if label == 'Year':
        xticklabel = [x for x in range(1994, 2004)]    
    elif label == 'Day of the Week':
        r = 0
        xticklabel = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun' ]
    elif label == 'Day of the Month':
        r = 90
        xticklabel = [x for x in range(1, 32)]
    else:
        r = 0
        xticklabel = [x for x in range(1, 13)]
    
    fig, ax = plt.subplots()
    ax.bar(bar_positions, bar_heights, .5, color=c, alpha=.6)
    ax.set_xticks(tick_position)
    ax.set_xticklabels(xticklabel, rotation=r)
    ax.set_xlabel(label)
    ax.set_ylabel('Total Births')
    ax.set_title('Total Births by ' + label + ' from 1994 to 2003')
    plt.show()
```

```python
transform_display(cdc_year_births, 'b','Year')
```

```python
transform_display(cdc_month_births, 'g','Month')
```

```python
transform_display(cdc_dom_births, 'y','Day of the Month')
```

```python
transform_display(cdc_dow_births, 'c','Day of the Week')
```

The yearly total births number over the year are stable although it drops to slightly lower level after 1998.
July and August has the highest births numbers.
There is not much difference from day to day of the month except 29, 30, 31 as not all the months have 29, 30, or 31 days.
Total Births number by days of the week shows interesting pattern.  Monday's total is higher than Saturday's or Sunday's. By Tuesday the total births number jumps to highest of the week.  Then the level drops a little and stays about the same for Wednesday, Thursday and Friday.  Total births number drop a lot more Saturday and Sunday as if babies know it is weekend.

### A function to calculate min and max values by different timeframe

```python
# Create a function that can calculate the min and max 
# values for any dictionry that's passed in

def find_min_max(data_dic):
    val_list = data_dic.values()
    return (min(val_list), max(val_list))
```

```python
# Create a function  that extracts the same values 
# across years and calculates the differences between 
# consecutive values to show if number of births is 
# increasing or decreasing.

def across_year(data, column, val):
    result = {}
    for row in data:
        year = row[0]
        if row[column] == val:
            if year in result:
                result[year] += row[4]
            else:
                result[year] = row[4]

# cdc births on Saturday from 1994 to 2003
cdc_births_saturday = across_year(cdc_list, 3, 6)
```
