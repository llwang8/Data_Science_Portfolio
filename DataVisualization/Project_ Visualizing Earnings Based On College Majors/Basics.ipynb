# Visualizing Earnings Based on College Majors
***

The dataset is about the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

The project aims to explore college major data on job outcomes using various plotting tools built into pandas. These tools include scatter plots to explore correlations between columns, pivot tables to aggregate data with certain criteria, histogram to examine distribution of columns and bar plots for comparison analysis. 

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

Rank - Rank by median earnings (the dataset is ordered by this column).
Major_code - Major code.
Major - Major description.
Major_category - Category of major.
Total - Total number of people with major.
Sample_size - Sample size (unweighted) of full-time.
Men - Male graduates.
Women - Female graduates.
ShareWomen - Women as share of total.
Employed - Number employed.
Median - Median salary of full-time, year-round workers.
Low_wage_jobs - Number in low-wage service jobs.
Full_time - Number employed 35 hours or more.
Part_time - Number employed less than 35 hours.

### Explore relations between columns using scatter plots
***


```python
fg, ax = plt.subplots()
```

```python
# Explore the relations between sample_size and median
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median Salary vs. Sample Size')
plt.show()
```

Median of salary of graduates has no obvious link with sample size.

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample_size vs. Unemployment Rate')
plt.show()
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full Time vs. Median of Salary')
plt.show()
```

Full time status has no obvious link with median of salary of graduates.

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Women as Share of Total Vs. Unemployment Rate')
plt.show()
```

Women share of total graduates has no obvious link with Unemployment rate.

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Male Graduate vs. Median Salary')
plt.show()
```

Male graduates number has not obvious link with median of graduates salary.

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women Graduate vs. Median Salary')
plt.show()
```

Number of women graduates has no obvious relation with median of graduates salary.

```python
ax = recent_grads.plot(x='Rank', y='Median', kind='scatter')
ax.set_title('Median of Salary vs. Ranking of Majors')
plt.show()
```

Median of salary has negative relation with ranking of major, which means that students make more money in more popular majors (rank number lower)

```python
ax = recent_grads.plot(y='Median', x='ShareWomen', kind='scatter')
ax.set_title('Women Share of Total vs. Median of Salary')
plt.show()
```

Median salary of graduates has weak nagative link to women share of total, which means that students that majored in subjects that were majority female make slightly less money than the opporsite situation.

### Use a grouped bar plot to compare the number of men with the number of women in each category of majors.
